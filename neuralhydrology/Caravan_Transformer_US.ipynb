```python
with open(run_dir / "test" / "model_epoch030" / "test_results.p", "rb") as fp:
    results = pickle.load(fp)
    
results.keys()
```

```python
import pandas as pd
metrics_results = []

for i in results.keys():
  qobs = results[str(i)]['1D']['xr']['streamflow_obs']
  qsim = results[str(i)]['1D']['xr']['streamflow_sim']

  values = metrics.calculate_all_metrics(qobs.isel(time_step=-1), qsim.isel(time_step=-1))
  values['Basin_ID'] = i

  metrics_results.append(values)
metrics_results = pd.DataFrame(metrics_results)
```

```python
metrics_results.to_csv('/content/drive/MyDrive/neuralhydrology/exp/Transformer_4head_4layer_Caravan_US.csv')
```
